## Amortecimento por atrito

O amortecimento por atrito seco (ou de Coulomb) ocorre devido às imperfeições das superfícies quando estas entram em contato. O valor da força de atrito é constante e diretamente proporcional a força normal que atua entre as superfícies.
<img src="atrito.png" alt="drawing" width="500"  />

\begin{align}
f_c = \mu N
\end{align}

Percebe-se, portanto, que o amortecimento de Coulomb independe da magnitude da velocidade do sistema, porém, atuará sempre no sentido contrário da mesma. Sendo assim, é possível definir que a força de atrito mudará de sinal de acordo com o sentido do movimento do sistema, podendo-se escrever a equação de equilíbrio dinâmico como:

\begin{align}
m\ddot{x} + kx = -\mu N
\end{align}

Quando o corpo se desloca para a direita e

\begin{align}
m\ddot{x} + kx = \mu N
\end{align}

Quando o corpo se desloca para a esquerda.

A equação de equilíbrio dinâmico é uma **Equação diferencial ordinária não homogênea**, em que a sua solução pode ser escrita como:

\begin{align}
x = x_{homogênea} + x_{particular}
\end{align}

A solução homogênea consiste em resolver a expressão $m\ddot{x} + kx = 0$, ou seja, coincide com a equação de vibração livre não amortecida.

\begin{align}
x_{homogênea} = A_1cos(w_nt) + A_2sen(w_nt)
\end{align}

A solução particular consiste em utilizar o teorema da preservação da frequência, ou seja, a resposta deve possuir a mesma frequência do carregamento externo. Como o carregamento externo é uma constante (frequência nula), a solução também será uma constante.

\begin{align}
x_{particular} = C
\end{align}

Substituindo C nas equações de equilíbrio dinâmico, é possível encontra seu valor: $C =\pm \frac{\mu N}{k} \$ e seus movimentos podem ser definidos pelas seguintes equações:

\begin{align}
x(t) = A_1cos(w_nt) + A_2sen(w_nt) - \frac{\mu N}{k}
\end{align}

quando o sistema vai para a direita e

\begin{align}
x(t) = A_3cos(w_nt) + A_4sen(w_nt) + \frac{\mu N}{k}
\end{align}

quando o sistema vai para a esquerda.

Definindo como condições iniciais como $x_0 = x(0) $ e $v_0 = \dot{x}(0) = 0$ e considerando que o movimento começa indo para a **esquerda**, é possível calcular as constantes para o primeiro meio ciclo do movimento.

\begin{align*}
x(0) = x_0 &= A_3cos(w_n0) + A_4sen(w_n0) + \frac{\mu N}{k} \\
x_0 &= A_3 + \frac{\mu N}{k} \\
A_3 &= x_0 -  \frac{\mu N}{k}
\end{align*}

\begin{align*}
\dot{x}(0) = v_0 &= -w_nA_3sen(w_n0) + w_nA_4cos(w_n0)  \\
v_0 &= w_nA_4 + \frac{\mu N}{k} \\
A_4 &= 0
\end{align*}

Desse modo,a equação de equilíbrio para o primeiro meio ciclo é:

\begin{align}
x(t) = \left( x_0 - \frac{\mu N}{k} \right)cos(w_nt) + \frac{\mu N}{k}
\end{align}

O primeiro meio ciclo é finalizado quando $t = \frac{\pi}{w_n}$. Aplicando este valor na equação do movimento, é possível definir o valor inicial do próximo meio ciclo, que ocorrerá com o sistema se deslocando para a **direita**.

\begin{align*}
x(\frac{\pi}{w_n})= x_1 &= \left( x_0 - \frac{\mu N}{k} \right)cos(w_n\frac{\pi}{w_n}) + \frac{\mu N}{k}\\
x_1 &= -\left( x_0 - \frac{\mu N}{k} \right)+ \frac{\mu N}{k}\\
x_1 &= -\left( x_0 - \frac{2\mu N}{k} \right)
\end{align*}

Aplicando este valor na equação do movimento para a direita, encontram-se as constantes:
\begin{align}
x_1 &= A_1cos(w_n\frac{\pi}{w_n}) + A_2sen(w_n\frac{\pi}{w_n}) - \frac{\mu N}{k}\\
x_1 &= -A_1 - \frac{\mu N}{k}\\
A_1 &= \left( x_0 - \frac{3\mu N}{k} \right)
\end{align}

Analogamente ao movimento para a esquerda, a componente $A_2$ está vinculada à velocidade do movimento. Como os pontos analisados são pontos de inversão, a velocidade nestes será sempre nula, portanto, $A_2 = 0$. Dessa forma, a equação do movimento do segundo meio ciclo pode ser expressa por:

\begin{align}
x(t) = \left( x_0 - \frac{3\mu N}{k} \right)cos(w_nt) - \frac{\mu N}{k}
\end{align}

Repetindo-se o mesmo procedimento para os demais ciclos, é possível plotar o gráfico de deslocamento em função do tempo para o sistema.
<img src="decremento.png" alt="drawing" width="950" />

O movimento prosseguirá desta forma, mudando de equação a cada meiociclo, até que no final de um determinado meio ciclo o deslocamento seja tão pequeno que a força da mola seja incapaz de vencer a força de atrito estático. Isso acontecerá no final do meio ciclo de ordem r, que pode ser determinado por:
\begin{align}
r = \pm \left[\frac{x_0-\frac{\mu N}{k}}{\frac{2 \mu N}{k}} \right]
\end{align}

## Amortecimento histerético

O amortecimento por histerese ocorre devido ao atrito interno entre as fibras do material que está vibrando. A força reativa devido ao amortecimento por histerese é proporcional ao deslocamento do sistema:

\begin{align}
f_c = ki\gamma x
\end{align}

Dessa forma, a equação de equilíbrio do sistema é dada por:

\begin{align*}
m \ddot{x} + ki\gamma x + kx = 0 \\
m \ddot{x} + k(i\gamma x + x) = 0
\end{align*}

Divindindo ambos os lados pela massa:

\begin{align}
\ddot{x} + w_n^2(i\gamma x + x) = 0 
\end{align}

Onde $k(i\gamma x + x)$ é chamada de rigidez complexa e $\gamma$ é o fator de amortecimento estrutural.

Solucionando a equação de equilíbrio, obtém-se a equação do movimento:
\begin{align}
x(t) = Ce^{-\xi_1 w_n t} sen(\xi_2w_nt+\phi)
\end{align}

em que:
\begin{align}
\xi_1 = \sqrt{\frac{\sqrt{1+\gamma^2}-1}{2}}\\
\xi_2 = \sqrt{\frac{\sqrt{1+\gamma^2}+1}{2}}
\end{align}

O decremento logarítimo neste caso é:

\begin{align}
\delta = ln \left( \frac{x_n}{x_{n+1}} \right) = \frac{2\pi\gamma}{\sqrt{1+\gamma^2}+1}
\end{align}

Quando $\gamma << 1$, $\sqrt{1+\gamma^2} \approx 1$, então se pode aproximar o decremento logarítmico por:

\begin{align}
\delta = \pi\gamma
\end{align}
